```python
import zipfile
import os

# Define the name for your output ZIP file
zip_filename = "all_files.zip"

# Create a ZIP file and add all files in the current directory
with zipfile.ZipFile(zip_filename, "w", zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk("."):
        for file in files:
            # Skip the zip file itself if it already exists
            if file == zip_filename:
                continue
            file_path = os.path.join(root, file)
            # Write file to the ZIP archive with relative path
            zipf.write(file_path, os.path.relpath(file_path, "."))

print(f"All files have been zipped into '{zip_filename}' successfully.")
```
