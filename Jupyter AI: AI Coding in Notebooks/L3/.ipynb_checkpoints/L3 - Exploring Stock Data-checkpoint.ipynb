- To find the dates that correspond to the peak and lowest prices, use a prompt like this:
  > Use the Close column of aapl dataFrame to find and print:
  > - the peak date (in a variable called `peak_date`) that corresponds to the maximum closing price `peak_price`
  > - the lowest date (in a variable called `lowest_date`) that corresponds to the minimum closing price `lowest_price`
  >
  > Update the above code to show the peak and low prices in the line chart.

```python
import matplotlib.pyplot as plt

# Find peak and lowest points
peak_price = aapl['Close'].max()
lowest_price = aapl['Close'].min()
peak_date = aapl['Close'].idxmax()
lowest_date = aapl['Close'].idxmin()

print(f"Peak Date: {peak_date}, Peak Price: {peak_price:.2f}")
print(f"Lowest Date: {lowest_date}, Lowest Price: {lowest_price:.2f}")

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(aapl['Close'], color='royalblue', linewidth=2, label='Closing Price')

# Highlight the peak and lowest points
plt.scatter(peak_date, peak_price, color='green', s=100, zorder=5, label='Peak Price')
plt.scatter(lowest_date, lowest_price, color='red', s=100, zorder=5, label='Lowest Price')

# Add annotations for better visualization
plt.annotate(f'Peak\n{peak_price:.2f}', 
             xy=(peak_date, peak_price), 
             xytext=(peak_date, peak_price * 1.02),
             arrowprops=dict(facecolor='green', shrink=0.05),
             ha='center')

plt.annotate(f'Low\n{lowest_price:.2f}', 
             xy=(lowest_date, lowest_price), 
             xytext=(lowest_date, lowest_price * 0.98),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center')

# Add title and labels
plt.title('Apple (AAPL) Closing Price Trend', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Adjust layout and display
plt.tight_layout()
plt.show()
```

- To find the context related to the peak and lowest date, use a prompt like this:
  > For the `peak_date` and `lowest_date`, search for related Apple news using Serper. The Serper API key is saved in a .env file. Store the snippets of the found articles in a json string `news_snippets` that has these fields: peak_date, lowest_date, peak_news_snippets, lowest_news_snippets.

<span style="color:green; font-weight:bold;">Note:</span> The `SERPER_API_KEY` variable is already defined in this environment, you do not need to create  an `.env` file. 

```python
import os
import json
import requests
from dotenv import load_dotenv

# Load the Serper API key from .env
load_dotenv()
SERPER_API_KEY = os.getenv("SERPER_API_KEY")

# Define your peak and lowest dates from the previous step
# (Assumes peak_date and lowest_date variables are already defined)
# Example:
# peak_date = '2022-01-03'
# lowest_date = '2022-06-16'

# Define a helper function for the Serper search
def search_news(query, api_key):
    url = "https://google.serper.dev/news"
    headers = {"X-API-KEY": api_key, "Content-Type": "application/json"}
    payload = {"q": query, "num": 5}  # fetch top 5 related articles
    response = requests.post(url, headers=headers, json=payload)
    response.raise_for_status()
    results = response.json().get("news", [])
    snippets = [item.get("snippet") for item in results if item.get("snippet")]
    return snippets

# Perform searches for the specific dates
peak_query = f"Apple news on {peak_date.date() if hasattr(peak_date, 'date') else peak_date}"
lowest_query = f"Apple news on {lowest_date.date() if hasattr(lowest_date, 'date') else lowest_date}"

try:
    peak_news_snippets = search_news(peak_query, SERPER_API_KEY)
    lowest_news_snippets = search_news(lowest_query, SERPER_API_KEY)
except Exception as e:
    print(f"Error during news search: {e}")
    peak_news_snippets, lowest_news_snippets = [], []

# Create the JSON string
news_snippets = json.dumps({
    "peak_date": str(peak_date),
    "lowest_date": str(lowest_date),
    "peak_news_snippets": peak_news_snippets,
    "lowest_news_snippets": lowest_news_snippets
}, indent=4)

# Print or save the result
print(news_snippets)
```

## Step 5: Analyze Volatility

- To calculate the signal's volatility, use a prompt like this: 
  >In the DataFrame aapl, find the overall volatility in percentage using the column Close. Volatility is the standard deviation of the daily percentage changes. Save the result in a variable called `volatility`.

```python
# Calculate daily percentage changes
daily_pct_change = aapl['Close'].pct_change()

# Calculate volatility (standard deviation of daily returns, converted to percentage)
volatility = daily_pct_change.std() * 100

print(f"Overall volatility: {volatility:.2f}%")
```

- To find and plot the rolling volatility, use a prompt like this:
  > Calculate the rolling volatility as the as 20-day standard deviation of the daily percentage change and plot it. Identify days of high volatility where volatility is greater than mean + std. Save the days of high volatility in a DataFrame called `high_vol_days`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate daily percentage changes
daily_pct_change = aapl['Close'].pct_change()

# Compute 20-day rolling standard deviation (rolling volatility in percentage)
aapl['Rolling_Volatility'] = daily_pct_change.rolling(window=20).std() * 100

# Plot the rolling volatility
plt.figure(figsize=(12, 6))
plt.plot(aapl['Rolling_Volatility'], label='20-Day Rolling Volatility', color='blue')
plt.title('AAPL 20-Day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')
plt.legend()
plt.grid(True)
plt.show()

# Determine high volatility threshold = mean + std
vol_mean = aapl['Rolling_Volatility'].mean()
vol_std = aapl['Rolling_Volatility'].std()
threshold = vol_mean + vol_std

# Identify high volatility days
high_vol_days = aapl[aapl['Rolling_Volatility'] > threshold]

# Display the DataFrame of high volatility days
high_vol_days.head()
```

## Step 6: Report Generation

- To generate a report summarizing the insights, use a prompt like this:
  
  > Use gpt-4.1-mini to generate a summary that takes in these variables:
  > - ticker: stock ticker (string)
  > - start_date: analysis starting period (string)
  > - end_date: analysis end period (string)
  > - numerical metrics: total_return & volatility (in percentage)
  > - peak_date, peak_price
  > - lowest_date, lowest_price
  > - high_vol_days: pandas DataFrame showing high volatility days
  > - news_snippets: string containing snippet of news for the peak and lowest dates
  >
  > The OpenAI API key is stored in the .env file. The variables are already defined in the notebook.

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

# Load the OpenAI API key from .env
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Prepare your prompt with the variables already defined in the notebook
prompt = f"""
Generate a concise financial summary for stock {ticker} from {start_date} to {end_date}.
Include:
- Total return: {total_return:.2f}%
- Overall volatility: {volatility:.2f}%
- Peak price: {peak_price:.2f} on {peak_date}
- Lowest price: {lowest_price:.2f} on {lowest_date}
- Notable high-volatility periods from the `high_vol_days` DataFrame
- Relevant news snippets: {news_snippets}

Summarize key insights about performance trends, volatility patterns,
and potential correlations with the given news.
Use a professional, analytical financial tone.
"""

# Call the gpt-4.1-mini model
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.7
)

# Extract the generative summary
summary = response.choices[0].message.content

print(summary)
```
